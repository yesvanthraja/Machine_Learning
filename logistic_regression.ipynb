```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn.linear_model import LogisticRegression
```

```python
data = pd.read_csv('dataset\Social_Network_Ads.csv')
data
```

```python
x = data.iloc[: ,[2,3]].values
y = data.iloc[:, 4].values
```

```python
from sklearn.model_selection import train_test_split
x_train, x_test, y_train, y_test =  train_test_split(x, y, test_size= 0.2, random_state=43)
```

```python
from sklearn.preprocessing import StandardScaler

sc = StandardScaler()
x_train  = sc.fit_transform(x_train)
x_test = sc.transform(x_test)
```

```python
x_train
```

```python
x_test
```

```python
logistic_reg = LogisticRegression()
logistic_reg.fit(x_train, y_train)
```

```python
logistic_reg.predict(x_test)
```

```python
print(f"Accuracy of the Logistic regression model is {(logistic_reg.score(x_train,y_train)) * 100} %")
```

```python
import gradio as gr
```

```python
def predict(age, salary):
    input_data = sc.transform([[age, salary]])
    prediction = logistic_reg.predict(input_data)
    return "Will Purchase the BMW car" if prediction[0] == 1 else "Will Not Purchase the BMW car"
```

```python
iface = gr.Interface(fn=predict, 
                     inputs=["number", "number"], 
                     outputs="text",
                     title="Logistic Regression Model for Purchase Prediction",
                     description="Enter Age and Estimated Salary to predict whether a person will purchase the BMW or not.")
```
